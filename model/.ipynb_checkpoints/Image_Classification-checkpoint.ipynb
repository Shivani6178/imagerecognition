```python
##### Drawing eye 
# roi - Region of interest

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex,ey,ew,eh) in eyes:
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)

plt.figure()
plt.imshow(face_img, cmap='gray')
plt.show()
```

```python
plt.imshow(roi_color, cmap="gray")
```

```python
def get_cropped_image(image_path):
    try:
        img = cv2.imread(image_path)
        
        if img is None:
            raise Exception("Error: Unable to read the image.")
        
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3, 5)
        
        if len(faces) == 0:
            raise Exception("Error: No face detected in the image.")
        
        for (x, y, w, h) in faces:
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = img[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray)
            if len(eyes) >= 2:
                return roi_color
        
        raise Exception("Error: No suitable face with eyes detected in the image.")
    
    except Exception as e:
        print(str(e))
        return None
 
```

```python
cropped_image = get_cropped_image('./test_images/messi.jpg')
cropped_image

if cropped_image is not None:
    plt.imshow(cropped_image)
    plt.show()
```

```python
path_to_data = "./dataset/"
path_to_cr_data = "./dataset/cropped/"
```

```python
import os
img_dirs = []
for entry in os.scandir(path_to_data):
    if entry.is_dir():
        img_dirs.append(entry.path)
```

```python
img_dirs
```

```python
# If Cropped folder exist remove it and then create it in dataset folder

import shutil

if os.path.exists(path_to_cr_data):
    shutil.rmtree(path_to_cr_data) # remove
os.mkdir(path_to_cr_data) # create
```

```python
"""
{
    'saniya_mirza':[
        './dataset/cropped/saniya_mirza/saniya_mirza1.png',
        './dataset/cropped/saniya_mirza/saniya_mirza2.png',
        './dataset/cropped/saniya_mirza/saniya_mirza3.png',
    ],
    'virat_kohli':[
        './dataset/cropped/virat_kohli/virat_kohli1.png',
        './dataset/cropped/virat_kohli/virat_kohli2.png',
        './dataset/cropped/virat_kohli/virat_kohli3.png',
    ]
}
"""
```

```python
# Convert images into cropped images

cropped_image_dirs = []
celebrity_file_names_dict = {}

for img_dir in img_dirs:
    count = 1
    name = img_dir.split('/')[-1]
    
    celebrity_file_names_dict[name] = []
    
    for entry in os.scandir(img_dir):
        roi_color = get_cropped_image(entry.path)
       
        if roi_color is not None:
            cropped_folder = path_to_cr_data + name
            if not os.path.exists(cropped_folder):
                os.makedirs(cropped_folder)
                cropped_image_dirs.append(cropped_folder)
                print("Generating cropped images in folder: ", cropped_folder)
                
            cropped_file_name = name + str(count) + '.png'
            cropped_file_path = cropped_folder + '/' + cropped_file_name
            
            cv2.imwrite(cropped_file_path, roi_color)
            celebrity_file_names_dict[name].append(cropped_file_path)
            
            count +=1
    
```

<h3> Feature Engineering </h3>

```python
import pywt

def w2d(img, mode="haar", level=1):
    imArray = img
    # Data Conversion
    # Convert to grascale
    imArray = cv2.cvtColor(imArray,cv2.COLOR_RGB2GRAY)
    # Convert to float
    imArray = np.float32(imArray)
    imArray /=255
    # Compute Coefficients
    coeffs = pywt.wavedec2(imArray, mode, level=level)
    
    # Process Coefficients
    coeffs_H = list(coeffs)
    coeffs_H[0] *= 0
    
    # Reconstructions
    imArray_H = pywt.waverec2(coeffs_H, mode)
    imArray_H *= 255
    imArray_H = np.uint8(imArray_H)
    
    return imArray_H
    
    
```

```python
im_har = w2d(cropped_image, 'db1', 5)
plt.imshow(im_har, cmap='gray')
```

```python
celebrity_file_names_dict
```

```python
class_dict = {}
count = 0

for name in celebrity_file_names_dict.keys():
    class_dict[name] = count
    count += 1
    
    
class_dict
```

```python
X = []
y = []

for name, training_files in celebrity_file_names_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        if img is None:
            continue
        scalled_raw_img = cv2.resize(img, (32, 32))
        img_har = w2d(img, 'db1', 5)
        scalled_img_har = cv2.resize(img_har, (32, 32))
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1), scalled_img_har.reshape(32*32,1)))
        X.append(combined_img)
        y.append(class_dict[name])
        
        
```
